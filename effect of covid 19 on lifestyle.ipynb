## **Objective:** Classification of individuals based on lifestyle changes

* Let's consider that a person has significantly changed their lifestyle if they have experienced an increase or decrease of more than 0.5 in at least one of the following variables: ***prod_inc, sleep_bal, new_skill, fam_connect, relaxed,*** and ***self_time***.

* We'll create a new binary variable named ***lifestyle_change*** where 1 represents a significant lifestyle change and 0 represents no significant lifestyle change.

```python
data = df.copy()
```

```python
# Create a new binary variable 'lifestyle_change'
change_columns = ['prod_inc', 'sleep_bal', 'new_skill', 'fam_connect', 'relaxed', 'self_time']
data['lifestyle_change'] = data[change_columns].apply(lambda row: np.any(np.abs(row) > 0.5), axis=1).astype(int)

# Display the first few rows of the dataset
data.head()
```

```python
# Check the balance of classes in 'lifestyle_change'
data['lifestyle_change'].value_counts(normalize=True)
```

# About 65% of the individuals in the dataset experienced a significant lifestyle change (lifestyle_change = 1), while the remaining 35% did not experience a significant lifestyle change (lifestyle_change = 0).

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['occupation'] = le.fit_transform(df['occupation'])
data['line_of_work'] = le.fit_transform(df['line_of_work'].astype(str))
data['prefer'] = le.fit_transform(df['prefer'])
data['certaindays_hw'] = le.fit_transform(df['certaindays_hw'])
data['age']=le.fit_transform(df['age'].astype(str))
data['gender']=le.fit_transform(df['gender'].astype(str))
```

```python
data
```

```python
#X=data.iloc[:,0:-1]#-1 it means we drop dependent variable that is target column
#y=data.iloc[:,-1]
X=data.drop('lifestyle_change',axis=1)
y=data.lifestyle_change
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=50)

print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()#feature scaling
sc.fit(X_train) 
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
classification_rbf = SVC(kernel='rbf')#Gaussian kernel is non linear kernel function.
classification_rbf.fit(X_train, y_train)
```

```python
accu_svc=classification_rbf.score(X_test, y_test)
accu_svc
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=100, criterion='gini')
classifier.fit(X_train, y_train)
y_pred = classifier.predict
```

```python
print("Random Forest Training Accuracy:", classifier.score(X_train , y_train))
print("Random Forest Testing Accuracy:", classifier.score(X_test , y_test))
```

```python
accu_rfc=classifier.score(X_test,y_test)
accu_rfc
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score_rf=cross_val_score(RandomForestClassifier(n_estimators=100),X,y,cv=5)
print(score_rf)
print("Avg:",np.average(score_rf))
```

```python
plt.figure(figsize=(12,6))
model_acc = [accu_svc,accu_rfc]
model_name = ['SVC','RandomForestClassifier']
sns.barplot(x=model_acc, y=model_name,palette='viridis')
```
